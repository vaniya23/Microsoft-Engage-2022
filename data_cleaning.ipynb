```python
def con_var_summary(x):
    return pd.Series([x.dtype
                      ,x.isnull().count()
                      , x.isnull().sum() ,x.mean() , x.sum() , x.var() ,x.std(), x.min() ,
                      x.quantile(0.01) ,x.quantile(0.05) ,x.quantile(0.1) ,x.quantile(0.25) ,
                      x.quantile( ) ,x.quantile(0.75) ,x.quantile(0.90) ,x.quantile(0.95) ,x.quantile(0.99) ,
                     x.quantile(1)]
                    ,index = ['data_type' ,'N' , 'nulls' , 'mean','sum ','VAR' , 'STD' , 'MIN'
                             ,'P1','P5','P10' , 'Q1' , 'Q2' , 'Q3' , 'P90' ,'P95' , 'P99' , 'MAX'])
```

```python
cars.rename(columns = {'Ex-Showroom_Price' : 'Ex-Showroom_Price (in Rs.)' , 'Displacement' : 'Displacement (in cc)',
            'Fuel_Tank_Capacity': 'Fuel_Tank_Capacity (in litres)' ,'Height' : 'Height (in mm)', 'Length' : 'Length (in mm)',
            'Width' : 'width (in mm)' ,'City_Mileage':'City_Mileage (km/litre)' ,
             'Highway_Mileage':'Highway_Mileage (km/litre)','ARAI_Certified_Mileage':'ARAI_Certified_Mileage (km/litre)',
           'ARAI_Certified_Mileage_for_CNG':'ARAI_Certified_Mileage_for_CNG (km/kg)','Kerb_Weight':'Kerb_Weight (kg)',
            'Ground_Clearance':'Ground_Clearance (in mm)' , 'Front_Track':'Front_Track (in mm)',
            'Rear_Track':'Rear_Track (in mm)','Wheelbase':'Wheelbase (in mm)','Boot_Space':'Boot_Space (in litres)',
            } , inplace = True)
```

```python
def cat_var_summary(x):
    return pd.Series([x.dtype
                      ,x.isnull().count()
                      , x.isnull().sum()]
                    ,index = ['data_type' ,'N' , 'nulls'])
```
